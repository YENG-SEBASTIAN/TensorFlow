```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Create simple dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Build model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# Compile model
model.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)

# Train
history = model.fit(X, y, epochs=200, verbose=0)

```

```python
# Predict
prediction = model.predict(np.array([10.0]))
print("Prediction for 6:", prediction)
```

```python
plt.plot(history.history['loss'])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Over Time")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Initialize parameters
w = 0.0
b = 0.0

# Hyperparameters
learning_rate = 0.01
epochs = 200

# For plotting loss
loss_history = []

# Gradient Descent Loop
for epoch in range(epochs):
    # Predictions
    y_pred = w * X + b

    # Compute loss (MSE)
    loss = np.mean((y - y_pred) ** 2)
    loss_history.append(loss)

    # Compute gradients
    dw = (-2/len(X)) * np.sum(X * (y - y_pred))
    db = (-2/len(X)) * np.sum(y - y_pred)

    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    # Optional: print every 20 epochs
    if epoch % 20 == 0:
        print(f"Epoch {epoch}: w={w:.4f}, b={b:.4f}, loss={loss:.4f}")

# Final parameters
print("\nFinal parameters:")
print(f"w = {w}, b = {b}")

# Make a prediction
x_new = 6.0
y_new = w * x_new + b
print(f"Prediction for {x_new}: {y_new}")

# Plot loss curve
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Over Time (Manual Gradient Descent)")
plt.show()
```
